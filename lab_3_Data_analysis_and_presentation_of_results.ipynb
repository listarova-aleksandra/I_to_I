**Решение задачи и оформление результатов.**

**Необходимо рассчитать метрику happiness для всего набора данных**

**Как считать эту метрику:**
1.   Группируем датафрейм по пользователю, чтобы понять, как долго каждый пользовательь слушал музыку.
2.  Считаем общее время прослушиванния музыки
3. Находим медиану для суммы прослушиваний по пользователю.
4. Расчитываем разность со стандартным значением метрики hapiness = 57.456 и с нынешним значением.
5. Делаем вывод о том, нравится ли пользователю использовать сервис.







```python
user_total_play = df.groupby('user_id')['total_play_seconds'].sum()
```

```python
current_happiness = user_total_play.median()
current_happiness
```

```python
standart_happiness = 57.456
```

```python
difference_level_of_happiness = current_happiness - standart_happiness
difference_level_of_happiness
```

```python
if difference_level_of_happiness < 0:
    print('Удовлетворенность пользователей сервером понизилась.')
elif difference_level_of_happiness > 0:
    print('Удовлетворенность пользователей сервисом повысилась.')
else:
    print('Удовлетворенность пользователей сервисом осталась неизменной.')
```
